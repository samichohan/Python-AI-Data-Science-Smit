```python
# Project ATM
class ATM:
    def __init__(self):

        #instance variable ---> A variable for which value is different for different
        self.pin=""
        self.balance=0
        self.menu()

    def menu(self):
        user_input = input("""
        Hello! How would you like to proceed?
        
        1. Enter 1 to create pin
        2. Enter 2 to create deposit
        3. Enter 3 to create withdraw
        4. Enter 4 to create balance
        5. Enter 5 to create exit
        
            """)
        if user_input =="1":
            self.create_pin()
            
        elif user_input =="2":
            self.deposit()
            
        elif user_input =="3":
            self.withdraw()

        elif user_input =="4":
            self.check_balance()

        else:
            print("Bye!")
            
    def create_pin(self): 
        self.pin=input('Enter your pin ')
        print("Pin set successfully")
    
    def deposit(self):
        temp=input('Enter your pin')
        if temp == self.pin:
            amount=int(input('Enter your amount'))
            self.balance=self.balance+amount
            print('Deposit Successful')
        else:
            print('Invalid Pin')

    def withdraw(self):
        temp=input('Enter your pin')
        if temp == self.pin:
            amount=int(input('Enter your amount'))
            if self.balance>amount:
                self.balance=self.balance-amount
            else:
                print('Insufficient Balance')
        else:
            print('Invalid Pin')

    def check_balance(self):
        temp=input('Enter your pin')
        if temp == self.pin:
            print(self.balance)
        else:
            print('Invalid Pin')
        
ubl=ATM()
print('Pin :',ubl.pin)
print('Balance :',ubl.balance)

# ubl.create_pin()
# ubl.deposit()
# ubl.withdraw()
# ubl.check_balance()
```

Encapsulation:

Encapsulation means wrapping variables and methods together in a class and hiding the
internal details, so users interact with the object only through defined methods, not by
directly modifying its data.

Use Case: 

```python
# Encapsulatin ----> make private the instances
class Employee:
    def __init__(self,name,salary):
        self.name=name     # public attribute
        self.salary=salary  # private attribute
        
        
```

```python
emp = Employee('Fedrick',50000)
```

You notice that now the instance which make us private are not showing, this is encapsulation.

```python
# Public ---> shows to everyone 
emp.name
```

```python
# Private ---> Shows error because of access restrictions
emp.__salary
```

```python
# why this behaviour of private project ---> because of the difference between
emp.__salary=10000
emp.__salary
```

Inheritance:

Inheritance is an Object-Oriented Programming (OOP) principle where a class (called the child/derived class) can acquire the properties (attributes) and 
behaviors (methods) of another class (called the parent/base class). 

Use Case:

```python
# Inheritance Example

# Parent class
class User:
    def login(self):
        print('Login')
        
    def register(self):
        print('Register')

# Child class ----> child class access all the things of parents class        
class Student(User):
    
    def enroll(self):
        print('Enroll')
        
    def review(self):
        print('Review')

```

```python
s1=Student()
```

```python
# s1 is the objecr of child class thqat can acceess their own attributes and methods

# child class methods 
s1.enroll()
s1.review()

# parent class methods 
s1.login()
s1.register()
```

```python
# Example of inheritance with super method
# Parent class
class Car():
    def __init__(self,seats,tyres,stairing,brake,horn,car_name):
        self.seats=seats
        self.tyres=tyres
        self.stairing=stairing
        self.brake=brake
        self.horn=horn
        self.car_name=car_name

    def speed(self):
        print(f"the {self.car_name} is running")

    def describe_car(self):    
        print(f'''
        Car Model : {self.car_name}
        Total Seats : {self.seats}
        Tyres : {self.tyres}
        Horn Sound : {self.horn}
        Stairing Side : {self.stairing}
        Break : {self.brake}
        ''')

# Child class
class Electric_car(Car):
    def __init__(self,make,model,year,engine,seats,tyres,stairing,brake,horn,car_name):
        # super() method
        super().__init__(seats,tyres,stairing,brake,horn,car_name)
        self.make=make
        self.model=model
        self.year=year
        self.engine=engine

    def describe_electric_car(self):
        print(f'''
        Company : {self.make}
        Model Name : {self.model}
        Year : {self.year}
        Engine Sepcification : {self.engine}
        ''')
        
el=Electric_car('Tesla','ES01','2025','Erc20',4,4,'Right','jix','light','jinusa')    
# parent methods
el.speed()
el.describe_car()

# child method
el.describe_electric_car()
```

Polymorphism:

Use Case:

```python
# Example of polymorphism
class Animal:
    def sound(self):
        print('Animal makes a sound')

class Dog(Animal):
    def sound(self):
        print('Dog barks')

class Cat(Dog):
    def sound(self):
        print('cat meows')

animals = [Dog(), Cat(), Animal()]
for a in animals:
    a.sound()
```

```python
# polymorphism Example
class car():
    def __init__(self,seats,tyres,stairing,brake,horn,car_name):
        self.seats=seats
        self.tyres=tyres
        self.stairing=stairing
        self.brake=brake
        self.horn=horn
        self.car_name=car_name

    def speed(self):
        print(f"The {self.car_name} is running ")

    def sound(self):
        print(f"{self.car_name} sound is very Loud")

    def describe_car(self):
        print(f"""
        Car Model : {self.car_name}
        Total Seats : {self.seats}
        Tyres : {self.tyres}
        Horn Sound : {self.horn}
        Stairing Side : {self.stairing}
        Break : {self.brake}""")       
```

```python
class Car:
    def __init__(self,brand,model):
        self.brand=brand
        self.model=model
    def move(self):
        print('Drive!')

class Boat:
    def __init__(self,brand,model):
        self.brand=brand
        self.model=model
    def move(self):
        print('Sail!')

class Plane:
    def __init__(self,brand,model):
        self.brand=brand
        self.model=model
    def move(self):
        print('Fly')

# Create objects of different Types
car1=Car('Ford','Mustang')
boat1=Boat('Ibiza','Touring 20')
plane1=Plane('Boeing','JF Thunder 17')

# Put the object in a list
vehicles=[car1,boat1,plane1]

# iterate through the list and call the move() method
for x in vehicles:
    print(x.brand,x.model)
    x.move()
```

Abstract Use Case:

```python
# abstract class ---> inherit the ABC class , and have minimum one abstracted method 
# abstracted method --> have decorater of abstract method , and don't have any inside

# abstracet class
from abc import ABC, abstractmethod

class BankApp(ABC): # Inherit ABC class
    
    def database(self):
        print('connected to database')

    @abstractmethod # decorator
    def security(self):
        pass

class MobileApp(BankApp):   # inherit bank class
    def mobile_login(self):
        print('login to mobile')

    def security(self):
        print('mobile Security')
```

```python
mob=MobileApp()
```

```python
mob.database()
```

```python
mob.mobile_login()
```

```python
mob.security()
```

```python
class Point:
    def __init__(self,x,y):
        self.x = x
        self.y = y

    def __add__(self,other):
        return Point(self.x + other.x,self.y + other.x)

p1=Point(2,3)
p2=Point(4,5)
result = p1 + p2 # Actually calls p1.__add__(p2)

print(p1.x,p1.y)
print(p2.x,p2.y)
print(result.x,result.y)
```

Override Magical Methods:

```python
# override magic method
class Car():
    def __init__(self,windows,doors,enginetype):
        self.windows=windows
        self.doors=doors
        self.enginetype=enginetype

    def __str__(self):
        return "The Object has been Initialized"
    
    def drive(self):
        print('The Person drives a car.')
        
c=Car(6,7,'Petrol')
```

```python
# override object with string methods
print(c)
```

```python
# sizeof method ---> 5he entire object consist of 16 bytes 
c.__sizeof__()
```

```python
# override magic method
class Car():
    def __init__(self,windows,doors,enginetype):
        self.windows=windows
        self.doors=doors
        self.enginetype=enginetype

    def __str__(self):
        return "The Object has been Initialized"

    def __sizeof__(self):
        return "The size of object in bytes"
    
    def drive(self):
        print('The Person drives a car.')
        
c=Car(8,9,'Gas')
```
