```python
# identity operator ------> used to compare 33the objects, not if they are equal , but if they are actually the same object, with the same men

a=3
b=3

#is operator
print(a is b)

#is not  operator
print(a is not b)

x="Hello"
y="World"

print(x is y)
print(x is not y)

x="Hello-world"
y="Hello-world"
print(x is y)
print(x is not y)

m=[1,2,3,4]
n=[1,2,3,4]
print(m is n)
print(m is not n)
```

```python
#  Membership operator ----> shows the presence of a value in a sequence, output (True or True)

x="Islamabad"

# in operators
print("I" in x)

#not in operator
print("s" not in x)

x=[1,2,3,4,5]
print(3 in x)
print(6 not in x)
```

```python
# Bitwise Operators ------> used to compare (binary) numbers:
# & operator (and)
print(6 & 3)

"""
The  & operator  compares each bit and set it to 1 if both are 1, otherwise it is set to 0:
6 = 0000000000000110
3 = 0000000000000011
--------------------
2 = 0000000000000010
====================

"""
```

```python
 # | operator (or)
print(6 | 3)


"""
The | operator compares each bit and set it to 1 , otherwise it is set to 0:

6 = 0000000000000110
3 = 0000000000000011
--------------------
7 = 0000000000000111
====================

"""
```

```python
# ~ operator (not)
print(~3)

"""

The ~ operator incverts each bit (0 becomes 1 and 1 becomes 0).

invert 3 becomes -4:
 3 = 000000000000011
-4 = 111111111111100

Actual _binary: "0000 0000 0000 0011"
Inverted_binary: "1111 1111 1111 1100"
bit_Length: 16
conversion_type: "Two's complement to decimal"

steps:
  - step: 1
    title: "Check sign bit"
    explanation:
    -"The most significant bit (MSB) is  1."
    -"This indicates the number is negative."



 -  step: 2
   title: "Find magnitude"
   process:
   - "invert all bits:"
   - "1111 1111 1111 1100 -> 0000 0000 0000 0011"
   - "Add 1 to the inverted bits:"

   - 0000 0000 0000 0011
    +0000 0000 0000 0001
    --------------------
    0000 0000 0000 0100
    -------------------
   - "The magnitude is 4."

- step: 3

  title:"Apply sign"
  explanation:
   - "Since the sign bit was 1,the number is negative."
   - "Final decimal value = -4"


result:
  decimal_value: _4
  verification:
    - "Two's complement rule: for a negative number, invert bit and add 1 to get magnitude." 
    - "Attach a negative sign to the magnitude."

"""  
```

```python
# if and else
a = 330
b = 200
if b > a:
    print("b is greater than a")

else:
    print("a is greater than b")
```

```python
#Elif
a = 33
b = 33
if b > a:
    print("b is greater than a")
elif a == b:
    print("a and b are equal")
else:
    print("a is greater than b")
```

```python
# if-else with and operator
a = 200
b = 33
c = 500
if a > b and c > a:
    print("Both condition are true")
```

```python
# if-else with Or operator
a = 200
b = 33
c = 500
if a > b or a > c:
    print("At least one of the condition is True")
```

```python
# Nested if else
x = 8

if x > 10:
    print("Above ten,")
    if x > 20:
        print("and also above 20!")
    else:
        print("but not above 20,")
else:
    print("Less than ten")
```
