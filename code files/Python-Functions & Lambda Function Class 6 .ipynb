```python
# Calculator:

def calculator(arr,task=''):
    #add,sub,mul,div,sqr,avg,even,odd
    if task=='add':
        total=0
        for add in arr:
            total=total+add
        return total

    elif task=='sub':
        total=0
        for sub in arr:
            total=sub-total #[2,3,4,5,6,7]
        return total

    elif task=='mul':
        total=1
        for mul in arr:
            total=total*mul
        return total

    elif task=='div':
        total=1
        for div in arr:
            total=div/total
        return total

    elif task=='avg':
        total=0
        for avg in arr:
            total=total+avg
        return total/len(arr)

    elif task=='even':
        evenlist=[]
        for even in arr:
            if even%2==0:
                evenlist.append(even)
        return evenlist

    elif task=='odd':    
        oddlist=[]
        for odd in arr:
            if odd%2==1:
                oddlist.append(odd)
        return oddlist

    else:
        return arr

print(calculator([2,3,4,5,6,7],'add'))   
print(calculator([2,3,4,5,6],'sub'))   
print(calculator([2,3,4,5,6,7],'mul'))   
print(calculator([2,3,5,6,],'div'))   
print(calculator([2,3,4,5,6,7],'exponent'))   
print(calculator([2,3,4,5,6,7],'avg'))   
print(calculator([2,3,4,5,6,7],'odd'))   
print(calculator([2,3,4,5,6,7]))   
```

Use Case of Lambda Functions

Higher Order Functions

```python
l = [11,14,21,23,56,78,45,29,28,27,15,9]

def return_sum(l):
    even_sum=0
    odd_sum=0
    div3_sum=0

    for i in l:
        if i%2==0:
            even_sum= even_sum+i
        if i%2==1:
            odd_sum=odd_sum+i
        if i%3==0:
            div3_sum=div3_sum+i
    return even_sum,odd_sum,div3_sum
return_sum(l)        
```

```python
#hIGHER Order functions:--?> Perform addition  on the condition
l = [11,14,21,23,56,78,45,29,28,27,15,9]

def return_sum(func,l):
    result=0

    for i in l:
        if func(i):
            result=result+i
    return result

x= lambda x: x%2==0
y= lambda x: x%2==1
z= lambda x: x%3==0

print(return_sum(x,l))
print(return_sum(y,l))
print(return_sum(z,l))
```
