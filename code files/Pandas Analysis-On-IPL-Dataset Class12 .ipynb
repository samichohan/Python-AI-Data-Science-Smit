Drop Duplicates:

```python
# A very powerfull function use to drop duplicates row by row from the data.
# for all data
df.drop_duplicates()
```

```python
# Single columns ----> drop duplicates on basis of city
df.drop_duplicates(subset=['city'])
```

```python
# m ultiple columns ---> drop duplicated on the basis of city and season occurances
df.drop_duplicates(subset=['city','season'])
```

```python
# who is the the winner of each ipl season
df.drop_duplicates('season',keep='last')[['season','winner']].sort_values('season')
```
