1.Continous Memory (Compact Storage)

```python
import numpy as np
```

```python
py_list = [1,2,3,4,5]
np_array = np.array([1,2,3,4,5])

print("Python list elements addresses:")
for x in py_list:
    print(id(x))

print("\nNumpy array element addresses:")
print(np_array.__array_interface__["data"])
```

2.Vectorization(No Loops in Py* )

```python
#Python list with normal for loop:
list_data = [1,2,3,4,5]
list_result = []
for x in list_data:
    list_result.append(x * 2)
print("List result:",list_result)

#Numpy array (vectorized)
np_data = np.array([1,2,3,4,5])
np_result = np_data * 2
print("Numpy result:",np_result)
```

Faster:

```python
import time
```

```python
#Faster -->  Numopy array operational speed much faster than list operations
 # Python List
py_list = list(range(1_000_000))
start = time.time()
py_result = []
for x in py_list: # loop aperations
    py_result.append(x * 2)
end = time.time()
print("Python list time:", end - start , "seconds")

# Numpy array
np_array = np.arange(1_000_000)
start = time.time()
np_result = np_array * 2  # vectorized operations
end = time.time()
print("Numpy array time:", end - start , "seconds")
```

Indexing & Slicing in Numpy: 

```python
#simple array
arr12 = np.arange(24)
print(arr12)

#indexing
print(arr12[7])
print(arr12[-1])

#slicing
print(arr12[7:14])
print(arr12[-6:-1])
```

```python
#Multidimenstional matrics 
arr13 = np.arange(24).reshape(6,4) # reshape : converts the array into given matrics
print(arr13)

#indexing
print(arr13[2]) # return row

#slicing
print(arr13[2:5])  # return rows

#slicing rows & columns --> arr(slice rows , slice column)
print(arr13[:,2]) # return rows & column
print(arr13[:,1:4]) # return rows & columns
print(arr13[4:6,2:4]) # return rows & columns
```

```python
# multiblock & multidimenstional matrics:
arr14 = np.arange(32).reshape(2,4,4) #(block,row,cloumn)
print(arr14)

#indexing
print(arr14[0]) # first block
print(arr14[1]) # second block

print(arr14[0][0]) #first  bloxk first row

print(arr14[1][0]) # second block first row
 
print(arr14[0][0][1]) # first clock row 1 col 2 
print(arr14[1][0][1]) # second block row 1 col 2

#Slicing
print(arr14[:][0][0]) # All blocks ,first row , first col
print(arr14[:][1][1]) # All blocks ,second row, second col

print(arr14[0][:][:]) # Entire first block
print(arr14[1][:][:]) # Entire second block

print(arr14[:][:][0]) # All blocks,all rows
print(arr14[:][:][1]) # All blocks,all rows 
```

Broadcasting:

```python
#Example:
arr1 = np.arange(9).reshape(3,3)
arr2 = np.arange(3).reshape(1,3)

print("Array 1:\n",arr1)
print("Array 2: ",arr2)

print("Broad casting:\n",arr1+arr2)
```

Rules For Broadcasting:

```python
# 1.if x=m &  y=n, operation will execute
a1 = np.arange(16).reshape(2,8)
a2 = np.arange(8,24).reshape(2,8)

print("a1 :\n",a1)
print("a2 :\n",a2)
print("Broadcasting :\n",a1+a2)
```

```python
# 2. if x=1 & y=n then also operations will execute (same correspondance)
a3 = np.arange(4).reshape(1,4)
a4 = np.arange(12).reshape(3,4)

print("a3 :\n",a3)
print("a4 :\n",a4)

print("Broadcasting :\n",a3+a4)

a5 = np.arange(4).reshape(4,1)
a6 = np.arange(12).reshape(4,3)

print("a5 :\n",a5)
print("a6 :\n",a6)

print("Broadcasting :\n",a5+a6)
```

```python
# 3. if x=1 & y!=n then operation will not execute
a7 = np.arange(3).reshape(1,3)
a8 = np.arange(16).reshape(4,4)

print("a7 :\n",a7)
print("a8 :\n",a8)
print("Broadcasting :\n",a7+a8)
```

```python
# 4. if (1,n) & (m ,1) then if will become (m,n) , operaiortn will execute
a9 = np.arange(4).reshape(1,4)
a10 = np.arange(4).reshape(4,1)

print("a9 :\n",a9)
print("a10 :\n",a10)
print("Broadcasting :\n",a9+a10)
```

```python
# 5.if x = 1 & y=1 , then operations will also executw
a11 = np.arange(1).reshape(1,1)
a12 = np.arange(16).reshape(4,4)

print("a11 :\n",a11)
print("a12 :\n",a12)

print("Broadcasting :\n",a11+a12)

```

```python
# 6. if they are of diffferent dimenstions 
a13 = np.arange(4)
a14 = np.arange(24).reshape(6,4)

print("a13 :\n",a13)
print("a14 :\n",a14)

print("broadcasting :\n",a13+a14)
```
