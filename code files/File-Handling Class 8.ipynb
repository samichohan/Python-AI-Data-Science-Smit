```python
# check the error type
try:
    f=open("sample.tct","r")
    print(f.read())
    print(m)
except Exception as e:
    print(e)
    print(e.with_traceback)
```

```python
# catching specific exception  ---> for multiple errors
try:
    m=5
    f=open("sample1.txt","r")
    print(f.read())
    print(m)
    print(5/2)
    L=[1,2,3]
    L[100]
except FileNotFoundError:
    print("file not found")
except NameError:
    print("vaiable not defined")
except ZeroDivisionError:
    print("can't divide by 0")
except Exception as e:
    print(e)
```

```python
# else --> the work you have to do after success of try, you add that in the else
try:
    f=open("sample.txt","r")
except FileNotFoundError:
    print("File Not Found")
except Exception:
    print("Some error Occured")
else:
    print(f.read())
```

```python
# finally ---> no matter what, finally will execute every time
# else
try:
    f=open("sample1.txt","r")
except FileNotFoundError:
    print("File Not Found")
except Exception:
    print("Something went wrong")
else:
    print(f.read())
finally:
    print("This will execute every time")
```

```python
# Example 
class Bank:
    
    def __init__(self,balance):
        self.balance = balance
    def withdraw(self,amount):
        if amount < 0:
            raise Exception("amount cannot be -ve")
        if self.balance < amount:
            raise Exception('Amount exceeded to balance')
        self.balance = self.balance - amount

obj = Bank(1000)
try:
    obj.withdraw(100)
except Exception as e:
    print(e)
else:
    print(obj.balance)
```
