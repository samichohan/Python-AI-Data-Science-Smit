Sets Rules

 Topics

```python
 #Create Sets
s={} #This will make dictionary , not set because python make dictionary as implicit type conversion.
print(type(s))
s1={1,} # add comma after one variable can make it dictionary
print(type(s1))
s2={10,20,30,40,50,60} # homogenous
s3={'abdullah',10,20.7,True,(5+6j)}

print(s1,s2,s3)

#list with type conversion
s6=set()
s7=set('ABDULLAH')

print(s6,s7)
```

```python
#multidimenstional set ---> Rule 4. Set  don't allow suitable datatypes, so set also do not allow itself as a mutable data types in itself
s4={1,2,3,{4,5}} #2D
```

```python
# Duplicate ---> Rule 1. Set not allow duplicates in itself
s8={2,3,3,6,4,8,6,1,4,3,5}
print(s8)
```

```python
# Rule 5. ---> Set are unordered data type so it set the elements according to hashing algorithm
s9={'Hello',2,5.7,True}
print(s9)
```

```python
# Access items in set ---> Rule 2: Set do not allow indexing and slicing 
s10={1,2,3,4,5,6}

print(s10[3])
print(s10[2:5])
```

```python
# Add items in set ---> Rule 3. Seta itself is a mutable data type
s11={1,2,3,4,5,6,7}

#add ---> can add only one item
s11.add(8)
print(s11)

#update ---> it accept any iterale like list, tuple,set.
# ({4,5} was not added as a whole single object, But instead its inner elements (4 and 5) were extracted and inserted  into the set.)
s11.update([9, 10],{11, 12},(13, 14))
print(s11)
```

```python
# Delete Sets

s={10,20,30,40,50,60,70,80,90}

# Remove ---> remove the element in set by the actual value the element
s.remove(90)
print(s)

#pop ----> Remove the last element
# Rule 2.Set do not allow indexing and slicing (so it pop last element on the bsis of hashing)
s15={1,2,3,4,5}
s15.pop()
print(s15)

#clear ---> clear all the element from set
s15.clear()
print(s15)

# del ---> delete entire set
del s

# Rule 2. Set do not allow indexing and slicing (se not allow delete on basis of indexing)
s1={10,20,30,40,50,60,70,80,90}
del s1[3]
```

```python
# operation on set
s1={1,2,3,4,5}

#loop
for i in s1:
    print(i)

#membership operators
2 in s1

#product ---> Not allow in sets because it makes duplicate, Rule 1. Set not allow duplicates in itself
print(s1*3)

#addition of sets also (concatination) not exist in sets, instead set have union functions
```

```python
# Functions on set
s1={1,2,3,4,5}

#len
print(len(s1))

#min
print(min(s1))

#max
print(max(s1))

#sum
print(sum(s1))

#sorted ----> convert set to list implicitly
print(sorted(s1))
print(sorted(s1,reverse=True))

#union ---> it makes a new set from another set without chnage in the original set
s1={1,2,3}
s2=[3,4,5]
s3=s1.union(s2)
print(f"union = {s3}")
print(s2)
print(s1)  #original remains same

#intersection ----> Return the same elements in both sets
s1={1,2,3}
s2={3,4,5}
s3=s1.intersection(s2)
print(f"Intersection = {s3}")
print(s2)
print(s1)

#difference ---> Return the elements that are in s1 but not in s2
s1={1,2,3}
s2={3,4,5}
s3=s1.difference(s2)
print(f"Difference = {s3}")
print(s2)
print(s1)
```

Topics

```python
#create dictionary
d1={} # Empty dictionary
print(type(d1))

d2={"name":"Abdul Rehamn","Roll No":123,"Field":"Computer Science"} #normal dictionary

#Dictionary with type conversion
d3=dict()
d4=dict([('name','Abdul Sami'),('Roll No',123),('Field','Computer Science')])

# Multidimenstional Dictionary
d5={"Name":"Abdullah","Roll No":1020,"Marks":{"M1":45,"M2":67,"M3":89}} # 2D

print(d1,d2)
print(d3,d4)
print(d5)
```

```python
#Rule3 ----> Keys should always be immutable, but values can be mutable or immutable
d4={(1,2,3):'Abdullah'} #It accept tuple as a key because tuple is immutable
print(d4)

d5={[1,2,3]:'Abdullah'} #it is not accepting list as a key because list is mutable
print(d5)
```

```python
#Rule 5, ----> Keys should be unique 
d6={'Name':'Abdullah','Name':"Ali"} # It update the previous value of the key instead of giving error
print(d6)
```

```python
# Accessing values from dictionary ---> Access values from dictionary can be possible throught calling key output(value)
d7={"name":"Abdul Rehamn","Roll No":123,"Field":"Computer Science"} 
print(d7["name"],d7['Roll No'],d7["Field"])

#Accessing values from 2d dictionary
d8={"Name":"Abdullah","Roll No":1020,"Marks":{"M1":45,"M2":67,"M3":89}} # 2D
print(d8["Marks"]["M3"])

#get ---> Fetch value of the key
print(d8.get('Marks'))
```

```python
#Edit values from dictionary -----> dictionary_name[key]=new_value
d7={"name":"Abdul Rehamn","Roll No":123,"Field":"Computer Science"} 
d7['name']='Ali'
print(d7)

#Edit values from 2d dictionary 
d8={"Name":"Abdullah","Roll No":1020,"Marks":{"M1":45,"M2":67,"M3":89}} # 2D
d8['Marks']['M2']=78
print(d8)
```

```python
#Add new item in dictionary

#Add new key, value pairs in dictionnary
d9={"name":"Abdul Rehamn","Roll No":123,"Field":"Computer Science"} 
d9['Age']=16
print(d9)

#Add new key, value pairs in 2D dictionary
d10={"Name":"Abdullah","Roll No":1020,"Marks":{"M1":45,"M2":67,"M3":89}} # 2D
d10['Marks']['M4']=99
print(d10)

#update ---> add or modify key value pairs in the dictionnary
student= {'name':"Ali","age":'22'}

# updating existing key
student.update({'age':23})
print(student)

# Adding new key
student.update({"grade":"A"})
print(student)

# Merging another dictionary
student.update({"roll_no":101,"section":"B"})
print(student)
```
