```python
# import pandas
import numpy as np
import pandas as pd
```

Creating Series From List:

```python
#string series:
country = ['Pakistan','India','China','America','Russia','Turkey']

#series have 2 things ----> index , value
series = pd.Series(country)
print(series)

# Note : pandas call the string data type as object
```

```python
# integer series:
runs = [21,46,57,82,100]
pd.Series(runs)
```

```python
# creating series with custom index
# values 
marks = [66,78,45,89]
# index
subjects = ['math','english','physics','urdu']
pd.Series(marks,index=subjects)
```

```python
# setting a name of series 
mark=pd.Series(marks,index=subjects,name='Abdul Sami Chohan')
print(mark)
```

Creating Series From Dictionary:

```python
#dictionary
marks = {'math':67,'urdu':90,'physics':72,'chemistry':56,'english':48,'islamiat':55,'social studies':78,'general Knowledge':94,'computer science':83}
marks_series=pd.Series(marks,name='Abdul Sami Chohan Marksheet')
marks_series
```

Series Methods:

```python
# head ----> return starting 5 rows of series (by default)
print(marks_series.head())

# can change with the number of rows given number 
print(marks_series.head(7))
```

```python
# tail -----> return last 5 rows of series (by default)
print(marks_series.tail())

# can chqnge with  the number of rows given number 
print(marks_series.tail(7))
```

```python
# sample ----> return one row randomly from series (by default )
print(marks_series.sample())
# can change with the number of sample given number
print(marks_series.sample(n=3))
```

```python
# value_counts ----> return total frequency of each value from descending to ascending
marks={'math':67,'urdu':90,'physics':72,'chemistry':56,'english':48,'islamiat':5}
marks=pd.Series(marks)
marks.value_counts()
```

```python
# sort_values ---> sort all values in ascending (by default)
print(marks_series.sort_values())

#change the order to descending
print(marks_series.sort_values(ascending=False))
```

Series indexing & Slicing:

```python
# integer indexing 
print(marks)

# positive indexing
print(marks[5])
#negative indexing
print(marks[-2])

#string indexing
print(marks['islamiat'])
print(marks['english'])
```

```python
# integer slicing 

#positive sicing
print(marks[2:5])
#negative sicing
print(marks[-6:-2])

#string sicing
print(marks['urdu':'chemistry'])
```

```python
# Fancy indexing -----> multiple indexing values
marks[[1,3,5,7]]
```

Editing Series:

```python
print(marks)

#cahnge values using integer index
marks[2] = 80
print(marks)

# change values using string index 
marks['math'] = 100
print(marks)
```

```python
# what if the called index item at exist ----> pandas add that as new item at the 
marks['commerce'] = 99
print(marks)
```

```python
# it is also possible with slicing 
marks[2:6]=[100,80,60,40]
print(marks)
```

```python
# it is also possible with fancy indexing 
marks[[2,4,6]]=[0,0,0]
print(marks)
```

Important Series Methods:

```python
# as type --->change data type
print(marks)
print(marks.astype('float'))
```

```python
# duplicate---> return true if dublicates .else False
print(marks.duplicated())

#total dublicates
print(marks.duplicated().sum())
```

```python
#drop_duplicates----> remove the duplicates values
marks=marks.drop_duplicates()
print(marks)
```

```python
#making new series
x=pd.Series([1,2,3,np.nan,5,6,7,np.nan,9,10,np.nan,np.nan])
print(x)
```

```python
# isnull ---> return true if null value else false
print(x.isnull())

#total sum
print(x.isnull().sum())
```

```python
#dropna ---> use to remove the null values
print(x.dropna())
```

```python
#fillna ---> use to fill nan values by yourown logic
print(x.fillna(x.mean()))
```

```python
#isin ---> return True if value available ,else false
print(x.isin([3]))

print(x.isin([3,8,9])) # more than one

#boolean indexing 
print(x[x.isin([3,8,9])])
```
