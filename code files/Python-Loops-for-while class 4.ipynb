while loop with conditions

```python
# Even numbers

print("Even numbers")

count = 0
while count < 20:
    if count%2==0:
        print(count)
    count += 1
```

```python
# odd numbers
print("odd numbers")
count = 0
while count < 20:
    if count%2==1:
        print(count)
    count += 1    
```

```python
# Program for guessing the number
import random

jackpot=random.randint(1,100)

guess=int(input("Enter the number: "))
counter = 1

while guess!=jackpot:
    if guess<jackpot:
        print("Guess Higher")
    else:
        print("Guess lower")
    guess=int(input("Enter the number: "))
    counter += 1

print("Write Answer")
print(f"Total Attempt: {counter}")
```

```python
#slicing -----> string (start,end,step)

# (0,n-1)

a= "Hello World"

# positive

print(a[0:7])
print(a[2:])
print(a[:4])
print(a[:])
print(a[::2])
```
