TYPE CONVERSIOn

you can convert from one tyoe to another

2 TYPE OF TYPE CONVERSION

. Implicit: when puthon automatically perform type conversion.
. Explicit: when Programmer itself perform type conversion.



```python
#implicit -----> python automatically convert the data type after addition
print(9+9.5) #int & float
print(5+6+7j) # int & complex
print(2.5+9+5j) # int & complex
```

```python
# Explicit ------> where python can't perform implicit type conversion

# float to int
print(int(7.8))

# int to float
print(float(9))

# string  to int
print(int('45'))

#int to string
print((str(45)))

#float to string
print(str(6.7))

#string to float
print(float('6.7'))

#int to bool
print(bool(0))

#bool to int
print(int(False))

#int to complex
print(complex(5))
```

Note; you cannot convert complex number into another number type.

```python
# not convertable

print(int('karachi'))
print(float('karachi'))
print(complex('karachi'))
```

Literals

Raw Data given ina variable

4 types of literals

. Numeric
. String
. Boolean
. Special

```python
#1.Numeric literals
a= 0b1010 #binary literals
b= 100 #decimal literals
c= 0o310 #octa literals
d= 0x12c #Hexa literal

#float literals
float_1= 100.5
float_2= 1.5e2
float_3= 4.6e-3

#complex literals
x= 2.4j

print(a,b,c,d)
print(float_1,float_2,float_3)
print(x, x.real,x.imag)
```

String literals

```python
string='I am Abdullah'
strings="I am 20 years old"
single_char="c"
multiline_str="""This is multi line string with more than one line code,""" # for multiline string
unicode_str=u"\U0001f602" # universally contains every type of unicode of char,labels,emojis etc.....
raw_str=r"raw \n string" # use when want to write html in python

print(string)
print(strings)
print(single_char)
print(multiline_str)
print(unicode_str)
print(raw_str)
```

```python
#OPERATORS

#operators are use to perform operations on variables and value.python has following operatos.

#. Arithmetic Operators (+,-,*,/,%,//,**)(a+b)
#. Comparison Operators (>,<,>=,<=,==,!=)(a>b)
#. Logical Operators (and,or,not)(True,False)(a and b)
#. Assignment Operators (=,+=,-=,*-,/=,%=,//=,**=)
#. Identity Operators (is,is not)(True,False)(a is b)
#. Membership Operators (in,not in)(True,False)(a in b)
#. Bitwise operators (& ,and)
```

```python
# Arthmatic operators

a=7
b=2

#addition
print(a+b)


#subtraction
print(a-b)

#multiplication
print(a*b)

#division
print(a/b)  #output in float by default

#percent
print(a%b)  #modulus

#floor division
print(a//b)  #integer division

#exponent
print(a**b)  #power
```

```python
#comparison operators
a=15
b=10
#greater than,less than
print(a>b)
print(a<b)

#greater than or equal to , less than or equal to
print(a>=b)
print(a<=b)

# Equal Equal to, Not Equal to
print(a==b)
print(a!=b)
```

```python
# logical operators
x=True
y=False

# And operators
print(x and y) # True,False = False

# Or operators 
print(x or y) # True,False = True

# Not operator
print(not x)
```
