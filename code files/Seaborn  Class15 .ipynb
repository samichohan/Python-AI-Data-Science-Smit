Figure Level Function ---> relplot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# for using by default dataset
import plotly.express as px
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# scatter plot --> axes level plot
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time',size='size')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
```

```python
# Line Plot
gap = px.data.gapminder()
temp_df = gap[gap['country'] == 'Pakistan']
temp_df
```

```python
# axes level function
sns.lineplot(data=temp_df,x = 'year',y = 'lifeExp')
```

```python
# using relplot
sns.relplot(data=temp_df,x = 'year',y = 'lifeExp',kind='line')
```

```python
# hue -> style
temp_df=gap[gap['country'].isin(['Brazil','Pakistan','Germany'])]
temp_df
```

```python
# hue
sns.relplot(kind='line',data=temp_df,x='year',y='lifeExp',hue='country')
```

```python
sns.lineplot(data=temp_df,x='year',y='lifeExp',hue='country',style='continent',size='continent')
```

```python
# facet Plot --> figure level function 
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex',row='day')
```

```python
# col wrap
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=3)
```

3.Matrix Plot

```python
# heatmap

#Plot rectangular date as a color -encoded matrix
temp_df = gap.pivot(index='country',columns='year',values='lifeExp')

# axes level Function
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
# annot & linewidth & CMAP
temp_df = gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')

plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True,linewidth=0.5,cmap='viridis')
```

```python
# clustermap
# Plot aa matrix dataset as a hierarchically-clusteered heatmap
# this function requires scipy to be available

iris = px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```

. A second way to plot Facet plots ---> FacetGrid

```python
# figure --> relplot ---> displot --> catplot -->limplot
sns.catplot(data=tips,x='sex',y='total_bill',kind='violin',col='day',row='time')
```

```python
g = sns.FacetGrid(data=tips,col='day',row='time',hue='smoker')
g.map(sns.boxplot,'sex','total_bill')
g.add_legend()
```

.Plotting Pairwise Relationship(PairGrid Vs Pairplot)

```python
sns.pairplot(iris,hue='species')
```

```python
g = sns.PairGrid(data=iris,hue='species')
g.map(sns.scatterplot)
```

```python
# map_diag ---> map_offdiag
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.kdeplot)
```

```python
# map_diag ---> map_upper --> map_lower
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
#vars
g = sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```
